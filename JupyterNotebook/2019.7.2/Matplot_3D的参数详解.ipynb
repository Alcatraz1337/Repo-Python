```python
# -*- coding: utf-8 -*-
"""
3d
"""

import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
# figure
fig = plt.figure()
# 3d
# figure3d
ax = Axes3D(fig)

#ax = fig.add_subplot(111, projection='3d')

# x, y
x = np.arange(-4, 4, 0.25)
y = np.arange(-4, 4, 0.25)

# 
X, Y = np.meshgrid(x, y)

# 
R = np.sqrt(X ** 2 + Y ** 2)
# Z
Z = np.sin(R)

# 3D


# rstride:  cstride:
# rcount:50ccount:  


# cmap
# from matplotlib import cm
# ax.plot_surface(X, Y, Z, rstride = 1, cstride = 1, cmap = cm.coolwarm)
# cmap = "rainbow" 
#  cmap,  rainbow 
#  rainbow  coolwarm, 
ax.plot_surface(X, Y, Z, rstride = 1, cstride = 14, cmap = plt.get_cmap('rainbow'))

# 3D,zdir  'z'|'x'|'y'| z,x,y
# zdir = 'z', offset = -2 z = -2
ax.contour(X, Y, Z, zdir = 'z', offset = -2, cmap = plt.get_cmap('rainbow'))

# zx,y
ax.set_zlim(-2, 2)

plt.show()
```
